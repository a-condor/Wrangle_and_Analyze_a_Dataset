## Cleaning Data
In this section, clean **all** of the issues you documented while assessing. 

**Note:** Make a copy of the original data before cleaning. Cleaning includes merging individual pieces of data according to the rules of [tidy data](https://cran.r-project.org/web/packages/tidyr/vignettes/tidy-data.html). The result should be a high-quality and tidy master pandas DataFrame (or DataFrames, if appropriate).

```python
# Make copies of original pieces of data

t_archive_clean = twitter_archive.copy()
t_image_predictions_clean = tweet_image_predictions.copy()
t_status_clean =tweet_status.copy()
```

### Issue #1: (9) p1, p2, p3 : dog name; in varying cases in tweet_image_predictions

#### Define: Replace the '_' with ' ' for values in each column

#### Code

```python
t_image_predictions_clean.p1 = t_image_predictions_clean.p1.str.replace('_', ' ')
t_image_predictions_clean.p2 = t_image_predictions_clean.p2.str.replace('_', ' ')
t_image_predictions_clean.p3 = t_image_predictions_clean.p3.str.replace('_', ' ')
```

#### Test

```python
t_image_predictions_clean.head()
```

### Issue #2: (11) Confidence columns should be in percentage for better understanding in tweet_image_predictions

#### Define: multiply p1,p2,p3_conf values by 100, convert  to an integer for whole number, then to string to append '%' sign

#### Code

```python
t_image_predictions_clean[['p1_conf', 'p2_conf', 'p3_conf']] = (t_image_predictions_clean[['p1_conf', 'p2_conf', 'p3_conf']] * 100).astype(int).astype(str) + '%'
```

#### Test

```python
t_image_predictions_clean[['p1_conf', 'p2_conf', 'p3_conf']].info()
```

```python
t_image_predictions_clean.head()
```

### Issue #3: (8) Error value  from text in denominator of constant 10 as shown in the describe function in twitter_archive

#### Define: Select only rows with text containing '**/10' as their rating using drop and contains. Ignore other rows without a standard rating format.

#### Code

```python
t_archive_clean.drop(t_archive_clean[~t_archive_clean.text.str.contains('/10')].index, inplace=True)
```

#### Test

```python
t_archive_clean.shape
```

```python
t_archive_clean.text.str.contains('/10').count()
```

### Issue #4: (6) Retweeted rows showing retweeted user id shouldn't be included in twitter_archive; users whose recored are in reply to another tweet are also deleted

#### Define: Select only rows containing the 'nan', this will ignore users that retweeted or commented on a post by using drop()

#### Code

```python
t_archive_clean.drop(t_archive_clean[~t_archive_clean.retweeted_status_user_id.astype(str).str.contains('nan')].index, inplace=True)
t_archive_clean.drop(t_archive_clean[~t_archive_clean.in_reply_to_status_id.astype(str).str.contains('nan')].index, inplace=True)
```

#### Test

```python
t_archive_clean.shape
```

### Issue #5: (5) Incorrect rating_numerator as noticed in index 2335; 9/10 instead of 8/10 in twitter_archive

#### Define: re-extract the values of the numerator ratings using extract function

#### Code

```python
t_archive_clean['rating_numerator'] = t_archive_clean.text.str.extract(r'(\d+/10)')
t_archive_clean['rating_denominator'] = t_archive_clean.text.str.extract(r'(\d+/10)')
t_archive_clean['rating_numerator'] = t_archive_clean['rating_numerator'].str.replace(r'/10', '')
t_archive_clean['rating_denominator'] = t_archive_clean['rating_denominator'].str.replace(r'(\d+/)', '')
```

#### Test

```python
t_archive_clean[['text', 'rating_numerator', 'rating_denominator']]
```

```python
(t_archive_clean['rating_denominator'] == 10).count()
```

### Issue #7: (3) Columns not needed and with missing data in twitter_archive and tweet_image_predictions

#### Define: drop columns that are not needed, or those with missing data 

#### Code

```python
t_archive_clean.drop(['in_reply_to_status_id', 'in_reply_to_user_id', 'retweeted_status_id', 'retweeted_status_user_id', 'retweeted_status_timestamp', 'source', 'expanded_urls', 
                      'name', 'doggo', 'floofer', 'pupper', 'puppo' ], axis=1, inplace=True)
```

```python
t_image_predictions_clean.drop(['jpg_url', 'img_num'], axis=1, inplace=True)
```

#### Test

```python
t_image_predictions_clean.info()
```

```python
t_status_clean.info()
```

### Issue #8: Columns with wrong data types in twitter_archive

#### Define: convert the timestamp to datetime value, and ratings to integers

#### Code

```python
t_archive_clean.timestamp = pd.to_datetime(t_archive_clean.timestamp)
t_archive_clean.rating_numerator = t_archive_clean.rating_numerator.astype(int)
t_archive_clean.rating_denominator = t_archive_clean.rating_denominator.astype(int)
```

### Issue #9: (13) One table is sufficient for representing our data, every other ones can be joined.

#### Define: rename the id of tweet_status to tweet_id. Merge the 3 dataframes on tweet_id to the left. Tweet id should  be a string

#### Code

```python
t_status_clean.rename(columns={'id': 'tweet_id'}, inplace=True)
```

```python
twitter_archive_master = t_archive_clean.merge(t_image_predictions_clean, on=['tweet_id']).merge(t_status_clean, on=['tweet_id'])
```

```python
twitter_archive_master.tweet_id = twitter_archive_master.tweet_id.astype(str)
```
